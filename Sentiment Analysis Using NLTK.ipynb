```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import emoji
from deep_translator import GoogleTranslator
from google.colab import drive
drive.mount('/content/drive')
```

```python
#NLTK VADER
nltk.download('vader_lexicon')
```

```python
data = ('/content/drive/MyDrive/Datasets CSV/Rahul Gandhi Instagram Comment.csv')
```

```python
df = pd.read_csv(data)
```

```python
#VADER initializing
sia = SentimentIntensityAnalyzer()
```

```python
#Emoji process and translate function
def process_and_translate(text):
  if pd.isna(text):
    return ""

  text = str(text)
  #Convert emoji into twext
  text_with_emoji = emoji.demojize(text, delimiters=(" ", " "))
  text_with_emoji = text_with_emoji.replace("_"," ") #remove underscore and give space

  #Translate from others language
  try:
    translated_text = GoogleTranslator(source='auto', target='en').translate(text_with_emoji)
    return translated_text
  except Exception as e:
    return text_with_emoji
```

```python
#Function for sentiment
def get_sentiment(text):
  if not text:
    return 'Neutral'

  #polarity score by VADER
  scores = sia.polarity_scores(text)

  if scores['compound'] >= 0.05:
    return 'Positive'
  elif scores['compound']  <= -0.05:
    return 'Negative'
  else:
    return 'Neutral'
```

```python
if 'Comment' in df.columns:
    print("Translating text and processing emojis... (This may take some time)")
    # Processing using the 'Comment' column
    df['processed_text'] = df['Comment'].apply(process_and_translate)

    print("Performing sentiment analysis...")
    df['sentiment'] = df['processed_text'].apply(get_sentiment)

    #  Printing results and saving to file ---
    print("\n--- Sentiment Summary ---")
    print(df['sentiment'].value_counts())

    '''output_file = 'advanced_sentiment_analysis.csv'
    df.to_csv(output_file, index=False)
    print(f"\nComplete results have been saved to the '{output_file}' file.")
else:
    print("\n[Error] No column named 'Comment' was found in your file. Your column names are:")
    print(df.columns.tolist())'''
```
